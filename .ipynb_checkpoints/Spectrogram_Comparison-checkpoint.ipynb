```python
folder = '/home/hashim/PHD/audio_data/AllAudioSamples/spectrogram_comparison/'
original_file = folder + 'Original.wav'
laser_file = folder + 'laser.wav'
synthetic_file = folder + 'synthesized.wav'
replayed_file = folder + 'replayed.wav'
```

```python
orig_audio, sr_orig = librosa.load(original_file, res_type='kaiser_fast')
orig_audio = librosa.util.normalize(orig_audio)

laser_audio, sr_laser = librosa.load(laser_file, res_type='kaiser_fast')
laser_audio = librosa.util.normalize(laser_audio)

synth_audio, sr_syn = librosa.load(synthetic_file, res_type='kaiser_fast')
synth_audio = librosa.util.normalize(synth_audio)

replayed_audio, sr_rep = librosa.load(replayed_file, res_type='kaiser_fast')
replayed_audio = librosa.util.normalize(replayed_audio)
```

```python
IPython.display.Audio(data=orig_audio, rate=sr_orig)
```

```python
IPython.display.Audio(data=laser_audio, rate=sr_laser)
```

```python
IPython.display.Audio(data=synth_audio, rate=sr_syn)
```

```python
IPython.display.Audio(data=replayed_audio, rate=sr_rep)
```

```python
stft_orig = librosa.stft(orig_audio, n_fft=256)
S_db_orig = librosa.amplitude_to_db(np.abs(stft_orig), ref=np.max)

stft_laser = librosa.stft(laser_audio, n_fft=256)
S_db_laser = librosa.amplitude_to_db(np.abs(stft_laser), ref=np.max)

stft_syn = librosa.stft(synth_audio, n_fft=256)
S_db_syn = librosa.amplitude_to_db(np.abs(stft_syn), ref=np.max)

stft_rep = librosa.stft(replayed_audio, n_fft=256)
S_db_rep = librosa.amplitude_to_db(np.abs(stft_rep), ref=np.max)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10,16))
# ax = axes.flatten()

img_orig = librosa.display.specshow(S_db_orig, x_axis='time', y_axis='log', sr=sr_orig, ax=axes[0])
axes[0].set(title='Original')
axes[0].label_outer()

img_laser = librosa.display.specshow(S_db_laser, x_axis='time', y_axis='log', sr=sr_laser, ax=axes[1])
axes[1].set(title='Laser')
axes[1].label_outer()


img_syn = librosa.display.specshow(S_db_syn, x_axis='time', y_axis='log', sr=sr_syn, ax=axes[2])
axes[2].set(title='Synthetic')
axes[2].label_outer()

img_rep = librosa.display.specshow(S_db_rep, x_axis='time', y_axis='log', sr=sr_rep, ax=axes[3])
axes[3].set(title='Replayed')
axes[3].label_outer()

fig.colorbar(img_laser, ax=axes, format="%+2.f dB")

fig.savefig('spectrogram_comparison.png')
```
